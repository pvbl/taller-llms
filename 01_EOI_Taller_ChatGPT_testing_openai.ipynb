# Test open-ai

```python
from getpass import getpass
from openai import OpenAI
import os
import json
from google.colab import userdata

client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
```

```python
response = client.completions.create(
        model="gpt-4o-mini",
        prompt = "La capital de España es..",
        max_tokens=250,
          )
```

```python
print(response.model_dump_json(indent=2))
```

```python
messages=[
            {"role": "system", "content": "Eres un experto en nutrición. Responde en menos de 100 palabras."},
            {"role": "user", "content": "¿Cuáles son algunas fuentes saludables de proteínas que puedo incluir en mi dieta?"},
        ]

response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages = messages,
        max_tokens=250,
          )
```

```python

print(response.model_dump_json(indent=2))
```

# Test Huggingface

```python
from transformers import pipeline
import torch

pipe = pipeline("text-generation", model="google/gemma-3-1b-it", device="cuda", torch_dtype=torch.bfloat16)

messages=[
            {"role": "system", "content": "Eres un experto en nutrición. Responde en menos de 100 palabras."},
            {"role": "user", "content": "¿Cuáles son algunas fuentes saludables de proteínas que puedo incluir en mi dieta?"},
        ]

output = pipe(messages, max_new_tokens=50)
output
```
